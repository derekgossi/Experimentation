```python
import numpy as np
import random
import pandas as pd

# Generate some toy multivariate gaussians
gaussian_1 = np.random.multivariate_normal([0, 0],[[5, 0], [0, 1]], 200)
gaussian_2 = np.random.multivariate_normal([5,5],[[1, 0], [0, 1]], 100)
gaussian_3 = np.random.multivariate_normal([-4, -4],[[1, 3], [0, 1]], 124)
gaussian_4 = np.random.multivariate_normal([-2, 7],[[1, 0], [0, 1]], 59)
gaussians = [gaussian_1, gaussian_2, gaussian_3, gaussian_4]

# Put them in a dataframe
for num, gaussian in enumerate(gaussians):
    df = pd.DataFrame({'x' : gaussian[:,0], 'y' : gaussian[:,1]})
    if num == 0: gaussian_data = df
    else: gaussian_data = pd.concat([gaussian_data, df], ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

# Plot the toy data
colors = ['r', 'b', 'y', 'g']

for gaussian, c in zip(gaussians, colors):
    plt.scatter(gaussian[:,0], gaussian[:,1], color=c)

plt.xlim([-12,12])
plt.ylim([-12,12])
plt.show()
```

```python
# Chinese Restaurant Process
def chinese_restaurant_process(num_customers, alpha):
    # Check if number of customers is positive
    if num_customers <= 0: return []
    
    # Initialize table assignments and the next open table
    table_assignments = []
    next_open_table = 1
  
    # Now generate table assignments for the rest of the customers.
    for i in range(1, (num_customers + 1)):
        # Customer n chooses to sit at a new table
        # with probability alpha / (alpha + n - 1)
        if random.random() < ( float(alpha) / (alpha + (i - 1)) ):
            # Customer sits at new table.
            table_assignments.append(next_open_table)
            next_open_table += 1
        
        # Customer chooses to sit at an existing table by choosing
        # a random seated customer and sitting at their table,
        # thus giving more weight to tables with lots of people already.
        else:
            which_table = table_assignments[random.randint(0,len(table_assignments)-1)]
            table_assignments.append(which_table)
  
    return table_assignments
```

```python
# Long run probabilities for CRP
def long_run_prob_CRP(num_customers, alpha, iters):
    # Initialize the number of customers sitting at each table to 0
    table_counts = [0] * num_customers
    
    # Do the CRP and keep adding table assignments to table_counts
    for iteration in range(iters):
        for i in chinese_restaurant_process(num_customers, alpha):
            table_counts[i-1] += 1
    
    # Convert table counts into PDF
    table_pdf = [ float(i) / ( num_customers * iters ) for i in table_counts ]
    return table_pdf
    
    

print long_run_prob_CRP(num_customers = 10, alpha = 0.1, iters = 1000)
```

```python
from sklearn.mixture import DPGMM

# Shuffle dataframe
gaussian_data = gaussian_data.reindex(np.random.permutation(gaussian_data.index))

# Choose a max number of components for the algorithm
max_components = 8

# Count the number of clusters the DPGMM chooses
num_clusters = []

# Try clustering at different sample sizes
for iteration in range(int(np.floor(len(gaussian_data) / 10)) - 2):
    # Number of samples to use
    max_sample_value = ((iteration + 2) * 10) 
    sample_set = gaussian_data[0:max_sample_value]
    
    # Fit Dirichlet Process Gaussian Mixture Model
    dpgmm_model = DPGMM(n_components=max_components)
    fitted_dpgmm = dpgmm_model.fit(sample_set)
    dpgmm_predictions = fitted_dpgmm.predict(gaussian_data)
    num_clusters.append(len(set(dpgmm_predictions)))
    
    # Append predicted labels to dataframe
    gaussian_data['predicted'] = dpgmm_predictions

# Give a unique color to each category
unique_categories = list(set(gaussian_data['predicted']))
color_labels = ['r', 'y', 'g', 'b', 'c', 'm', 'k', 'w']
colors = [color_labels[unique_categories.index(i)] for i in gaussian_data['predicted']]

# Plot predicted data
plt.scatter(gaussian_data['x'], gaussian_data['y'], c=colors)
plt.xlim([-12,12])
plt.ylim([-12,12])
plt.show()

print "Params " + str(fitted_dpgmm.get_params())
```
